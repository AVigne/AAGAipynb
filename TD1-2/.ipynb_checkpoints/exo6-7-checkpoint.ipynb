```python
#question 1
def gen_congru(a,c,m,seed):
    return(((a*seed)+c)%m)
```

```python
#question 2 
s = 0
for i in range(10):
    s = gen_congru(25214903917,11,2**48,s)
    print(s)
```

```python
#question 3
import matplotlib.pyplot as plt

un12 = [1]
un7 = [1]
un51 = [1]

for i in range(101):
    un12.append(gen_congru(12,0,101,un12[-1]))
    un7.append(gen_congru(7,0,101,un7[-1]))
    un51.append(gen_congru(51,0,101,un51[-1]))
```

```python
fig, ax = plt.subplots()
ax.scatter(un12[:-1], un12[1:])
plt.xlabel('Un')
plt.ylabel('Un + 1')
```

```python
fig, ax = plt.subplots()
ax.scatter(un7[:-1], un7[1:])
plt.xlabel('Un')
plt.ylabel('Un + 1')
```

```python
fig, ax = plt.subplots()
ax.scatter(un51[:-1], un51[1:])
plt.xlabel('Un')
plt.ylabel('Un + 1')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(un12[0:-2], un12[1:-1], un12[2:])
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(un7[0:-2], un7[1:-1], un7[2:])
```

```python
#question 2
def gen_java(seed):
    a = 25214903917
    c = 11
    m = 2**48
    res = (((a*seed)+c)%m)
    res0 = res >> 16
    if res0 >= 2**31:
        res0 = res0 - (2**32)
    return res,res0
    
    
```

```python
s = 156079716630527
for i in range(10):
    s,g = gen_java(s)
    print(g)
```
